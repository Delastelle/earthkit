```python
from geopy import distance

# (lat, lon)

def get_bounds(ca, cb):
    c_mn = (min(ca[0], cb[0]), min(ca[1], cb[1]))
    c_mx = (max(ca[0], cb[0]), max(ca[1], cb[1]))
    return c_mn, c_mx
    
pnt_upb = ( 37.789733, -122.402614 )
pnt_downb = (  37.784409, -122.394974 )
bnds = get_bounds(pnt_upb, pnt_downb)

print(bnds)
```

```python
from streetview import search_panoramas

def get_wh(bnds):
    ca, cb = bnds
    return distance.distance((ca[0], cb[1]), (cb[0],cb[1])).km, distance.distance((ca[0], ca[1]), (ca[0],cb[1])).km

def get_samples(bnds, interval: float):
    w,h = get_wh(bnds)
    ca, cb = bnds
    sample_cnt = (w//interval)*(h//interval)
    input(f"sampling {sample_cnt} locations, proceed?")
    sampled = set()
    for i in range(int(w//interval)):
        lt = ca[0]+i*interval
        for j in range(int(h//interval)):
            lw = ca[1]+j*interval
            sampled.add((lt,lw))
    return sampled

smp = get_samples(bnds, 0.05)
smp
```

```python
sampled_coords = list(map(lambda x: {'lat':x[0], 'lng':x[1]}, smp))
sampled_coords
```

```python
import pydeck as pdk
layer = pdk.Layer(
    "ScatterplotLayer",
    sampled_coords,
    get_position=['lng', 'lat'],
)
view_state = pdk.ViewState(latitude= (pnt_upb[0] + pnt_downb[0])/2, longitude= (pnt_upb[1]+pnt_downb[1])/2, zoom=10)
r = pdk.Deck(layers=[layer], initial_view_state=view_state)
r.show()
```
